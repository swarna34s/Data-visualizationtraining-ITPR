```python
#convert list to tuple and remove duplicates
list=[1,2,2,3,4,4]
result=tuple(set(list))
print(result)
```

```python
#find minimum and maximum in tuple
t=(5,1,8,3)
max_val=t[0]
min_val=t[0]

for i in t:
  if i>max_val:
    max_val=i
  if i <min_val:
    min_val=i

 print("max =",max_val)
 print("min =",min_val)

```

```python
#set difference and intersection
set1={1,22,3,4}
set2={3,4,5,6}

intersection=set1&set2
difference=set1-set2

print("intersection:",intersection)
print("difference:",difference)
```
