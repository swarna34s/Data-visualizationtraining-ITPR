```python
# Find lines present in file1 but not in file2

with open("file1.txt", "r") as f1, open("file2.txt", "r") as f2:
    lines1 = f1.readlines()
    lines2 = f2.readlines()

for line in lines1:
    if line not in lines2:
        print(line.strip())
```
