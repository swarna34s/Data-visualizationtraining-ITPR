```python
# Program to remove duplicate lines and preserve order

seen = set()

with open("input.txt", "r") as infile, open("output.txt", "w") as outfile:

    for line in infile:
        if line not in seen:
            outfile.write(line)
            seen.add(line)

print("Duplicate lines removed successfully!")
```
