```python
category_sales = {}
max_sale = 0
top_product = ""

try:
    file = open("sales.txt", "r")

    for line in file:
        try:
            data = line.strip().split(",")

            if data[0] == "product":
                continue

            product = data[0]
            category = data[1]
            price = float(data[2])
            quantity = int(data[3])

            total = price * quantity

            # Category total
            if category in category_sales:
                category_sales[category] += total
            else:
                category_sales[category] = total

            # Highest selling product
            if total > max_sale:
                max_sale = total
                top_product = product

        except:
            print("Invalid row skipped:", line)

    file.close()

    print("Sales per Category:", category_sales)
    print("Top Product:", top_product, max_sale)

except FileNotFoundError:
    print("File not found!")
```
