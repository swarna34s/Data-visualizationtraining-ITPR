```python
votes = [
    ("A", "user1"),
    ("B", "user2"),
    ("A", "user1"),  # duplicate vote
    ("C", "user3"),
    ("A", "user4"),
    ("B", "user5")
]

voted_users = set()
vote_count = {}

for candidate, user in votes:

    # Check duplicate voter
    if user in voted_users:
        print("Duplicate vote ignored from:", user)
        continue

    # Add to set
    voted_users.add(user)

    # Count vote
    if candidate in vote_count:
        vote_count[candidate] += 1
    else:
        vote_count[candidate] = 1

# Find winner
winner = max(vote_count, key=vote_count.get)

# Output
print("\nVote Count:", vote_count)
print("Winner:", winner)
```
