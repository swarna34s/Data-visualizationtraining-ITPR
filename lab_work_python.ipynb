```python
class InvalidPasswordException(Exception):
    pass

def validate_password(password):
    if len(password) < 8:
        raise InvalidPasswordException("Password must be at least 8 characters")

    has_upper = False
    has_lower = False
    has_digit = False

    for ch in password:
        if ch.isupper():
            has_upper = True
        elif ch.islower():
            has_lower = True
        elif ch.isdigit():
            has_digit = True

    if not (has_upper and has_lower and has_digit):
        raise InvalidPasswordException("Password must contain uppercase, lowercase, and digit")

    print("Valid Password")

# Example
try:
    validate_password("Test1234")
except InvalidPasswordException as e:
    print(e)
```
