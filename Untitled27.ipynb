```python
# Program to extract students with marks > 75

with open("student.txt", "r") as file:
    with open("result.txt", "w") as newfile:

        for line in file:
            # Remove extra spaces/newline
            line = line.strip()

            # Split data (name,marks,city)
            data = line.split(",")

            name = data[0]
            marks = int(data[1])
            city = data[2]

            # Condition check
            if marks > 75:
                newfile.write(line + "\n")

print("Data extracted successfully!")
```
