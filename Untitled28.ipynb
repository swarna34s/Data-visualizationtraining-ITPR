```python
import string

# Open file
with open("article.txt", "r") as file:
    text = file.read()

# Convert to lowercase
text = text.lower()

# Remove punctuation
for p in string.punctuation:
    text = text.replace(p, "")

# Split into words
words = text.split()

# Dictionary to store frequency
freq = {}

for word in words:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1

# Display result
for word, count in freq.items():
    print(word, ":", count)
```
